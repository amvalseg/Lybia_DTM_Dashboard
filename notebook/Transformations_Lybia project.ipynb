# Reading the file

```python
import pandas as pd

# Path to the Excel file
file_path = r"C:\Users\avalcarcels\Documents\Angela\Analisis_datos\Projects\Lybia_DTM\data\DTM_Libya_R54_Migrant_Dataset_PUBLIC.xlsx"

# Read a specific sheet (e.g., 'Sheet1')
sheet_name = 'Muhalla Main'  
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Display the first rows of the DataFrame
df.head()
```

# Headers and Index

```python
# Convert the first row into headers
df.columns = df.iloc[0]  # Assign the first row as headers
df = df[1:].reset_index(drop=True)  # Remove the first row and reset the index

# Verify the result
df.head()
```

```python
# Add an index column
df['Index'] = df.index
```

# Dataset Creation and Unpivot Transformations

In this part of the notebook, we will focus on dividing the initial dataset into different datasets that will later allow us to perform dimensional modeling. 

```python
# Display the columns of the original dataset to identify which columns will be included in each specific dataset.
df.columns
```

```python
# The columns that will not be used are removed.
columnstodrop = ['ADM4_Muhalla_AR', 
                  'Muhalla_pcode',
                  'Migrants in Muhalla',
                  'Migrant IND']
df = df.drop(columns=columnstodrop)
```

## 2. df_shelter: Creating the dataset with the accommodation-related columns along with the index.

```python
df_shelter = df[['Rented Accomodation (Self Paid) IND',
                 'Rented Accomodation (Paind by others) IND',
                 'Rented Accomodation (Paind by employer) IND',
                 "Squatting on other people's properties IND", 
                 'Abondoned buildings IND',
                 'Sheltered in their workplace IND',
                 'schools or other public buildings (IND)',
                 'In informal camp setting (IND)', 'No accomodation IND',
                 'Other Shelter', 
                 'Index']]
```

```python
df_shelter.columns
```

### 2.1. Reshape and Unpivot columns

```python
#By applying melt, the values from the decision columns are combined into a single column called "Shelter Settings"

df_shelter = pd.melt(
    df_shelter,
    id_vars=['Index'],
    value_vars=['Rented Accomodation (Self Paid) IND',
       'Rented Accomodation (Paind by others) IND',
       'Rented Accomodation (Paind by employer) IND',
       "Squatting on other people's properties IND", 
       'Abondoned buildings IND',
       'Sheltered in their workplace IND',
       'schools or other public buildings (IND)',
       'In informal camp setting (IND)', 'No accomodation IND',
       'Other Shelter'],
    var_name='Shelter Settings',
    value_name='Value'
)
```

```python
df_shelter
```

### 2.2. Transforming Values

```python
# Check unique values in the 'Shelter settings' column to identify potential transformations
df_shelter["Shelter Settings"].unique()
```

```python
# Dictionary containing the replacements to be made
replacements = {
        'Rented Accomodation (Self Paid) IND': 'Self Paid Accomodation',
        'Rented Accomodation (Paind by others) IND': 'Paid by Others Accomodation',
        'Rented Accomodation (Paind by employer) IND': 'Paid by Employer Accomodation',
        "Squatting on other people's properties IND": 'Squatting',
        'Abondoned buildings IND': 'Abandoned buildings',
        'Sheltered in their workplace IND': 'Workplace',
        'schools or other public buildings (IND)': 'Schools or other public buildings',
        'In informal camp setting (IND)': 'Informal camp setting'
    }
    
# Replace values in the 'Shelter settings' column using the dictionary
df_shelter['Shelter Settings'] = df_shelter['Shelter Settings'].replace(replacements)
    
```

### 2.3. Confirm it's ok an Save it

```python
df_shelter["Shelter Settings"].unique()
```

## 3. df_shelter: Creating the dataset with the arrival time-related columns along with the index.

```python
df_arrivaltime = df[['Migrants arrived last month',
       'Migrants arrived between 1 to 6 months ago',
       'Migrants arrived between 6 to 12 months ago',
       'Migrants arrived Over 12 months ago',
       'Migrants arrival time unknown',
       'Index']]
df_arrivaltime
```

### 3.1. Reshape and Unpivot columns

```python
df_arrivaltime = pd.melt(
    df,
    id_vars=['Index'],
    value_vars=['Migrants arrived last month',
       'Migrants arrived between 1 to 6 months ago',
       'Migrants arrived between 6 to 12 months ago',
       'Migrants arrived Over 12 months ago',
       'Migrants arrival time unknown'],
    var_name='Arrival Time',
    value_name='Value'
)
df_arrivaltime
```

### 3.2. Transforming Values

```python
df_arrivaltime["Arrival Time"].unique()
```

```python
df_arrivaltime["Arrival Time"] = df_arrivaltime["Arrival Time"].replace({
    'Migrants arrived last month': "Last month",
    'Migrants arrived between 1 to 6 months ago':"1 to 6 months ago",
    'Migrants arrived between 6 to 12 months ago':"6 to 12 months ago",
       'Migrants arrived Over 12 months ago':"Over 12 months ago",
       'Migrants arrival time unknown': "Unknown"
})
```

### 3.3. Confirm it's ok an Save it

```python
df_arrivaltime
```

```python
df_arrivaltime.to_csv(r"filepath.csv")
```

## 5. df_demographics: Creating the dataset with the demographics-related columns along with the index.
